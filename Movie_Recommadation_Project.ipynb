```python
import pandas as pd
from google.colab import files

print("ðŸ“‚ Upload movies.csv")
movies_upload = files.upload()

print("ðŸ“‚ Upload ratings.csv")
ratings_upload = files.upload()

movies = pd.read_csv(list(movies_upload.keys())[0])
ratings = pd.read_csv(list(ratings_upload.keys())[0])


avg_ratings = ratings.groupby("movieId")["rating"].mean().reset_index()
avg_ratings.rename(columns={"rating": "avg_rating"}, inplace=True)

movies_data = pd.merge(movies, avg_ratings, on="movieId", how="left")
movies_data["avg_rating"] = movies_data["avg_rating"].fillna(0)


def get_unique_genres():
    all_genres = set()
    for g in movies_data["genres"]:
        if pd.isna(g):
            continue
        for genre in g.split("|"):
            all_genres.add(genre)
    return sorted(list(all_genres))

def get_movies_by_genre(genre, top_n=10):
    # Filter movies containing the genre
    genre_movies = movies_data[movies_data["genres"].str.contains(genre, case=False, na=False)]

    # Sort by rating
    genre_movies = genre_movies.sort_values("avg_rating", ascending=False)

    return genre_movies[["title", "avg_rating", "genres"]].head(top_n)


print("\nðŸŽ­ Do you want to see all genres? (yes/no)")
choice = input("Enter here: ").strip().lower()

if choice == "yes":
    print("\nðŸŽ­ Available Genres:")
    print(get_unique_genres())

print("\nðŸŽ¬ Enter a genre to get top movies (example: Action, Comedy, Drama):")
user_genre = input("Enter genre: ").strip()

print(f"\nðŸŽ¬ Top Movies in Genre: {user_genre}\n")
movies_list = get_movies_by_genre(user_genre)

print(movies_list.to_string(index=False))
```
